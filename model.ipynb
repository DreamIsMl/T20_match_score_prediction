```python
import pandas as pd
import pickle
```

```python
df = pickle.load(open('dataset_level2.pkl','rb'))
```

```python
df[df['city'].isnull()]['venue'].value_counts()
```

```python
import numpy as np
df['city'] = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
df['city'].isnull().sum()
```

```python
eleigible_city= df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
df = df[df['city'].isin(eleigible_city)]
```

```python
df.shape
```

```python
df.head()
```

```python
import pandas as pd

# Assuming df is your DataFrame with a 'runs' column and a 'match_id' column

# Convert 'runs' column to numeric type
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')  # Use errors='coerce' to handle non-numeric values gracefully

# Create 'current_score' column using cumsum
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df.head()
```

```python
df.drop(['venue'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split('.')[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
```

```python
df.head()
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
```

```python
df['balls_left'] = 120 - df['balls_bowled']
```

```python
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
import pandas as pd

# Assuming df is your DataFrame with a 'player_dismissed' column and a 'match_id' column

# Convert 'player_dismissed' column to numeric type
df['player_dismissed'] = pd.to_numeric(df['player_dismissed'], errors='coerce')  # Use errors='coerce' to handle non-numeric values gracefully

# Create 'player_dismissed_cumsum' column using cumsum
df['player_dismissed_cumsum'] = df.groupby('match_id')['player_dismissed'].cumsum()

# Calculate 'wicket_left' column
df['wicket_left'] = 10 - df['player_dismissed_cumsum']
```

```python
df.head()
```

```python
import pandas as pd

# Assuming df is your DataFrame

groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []

for id in match_ids:
    last_five.extend(groups.get_group(id)['runs'].rolling(window=30, min_periods=1).apply(sum).tolist())

# Now last_five should contain the desired values
```

```python
len(last_five)
```

```python
df['last_five'] = last_five
```

```python
df
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df.head(1)
```

```python
final_df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wicket_left','crr','last_five','runs_x']]
```

```python
df.head()
```

```python
final_df['last_five'].describe()
```

```python
final_df.drop(['crr'],axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['crr'] = final_df['crr'].fillna(final_df['crr'].median())
```

```python
final_df['last_five'].unique()
```

```python
mean_last_five = final_df['last_five'].mean()
median_last_five = final_df['last_five'].median()

def fill_null(row):
    value = row['last_five']
    if 0 <= value <= 15:
        return (0.5 * mean_last_five) + (0.5 * median_last_five)
    else:
        return value

final_df['last_five'] = final_df.apply(fill_null, axis=1)
```

```python
final_df.head()
```

```python
df.head()
```

```python
final_df
```

```python
x = final_df.drop(['runs_x'],axis=1)
y = final_df['runs_x']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=.2,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,BaggingRegressor,AdaBoostRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.linear_model import Ridge,Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
xtrain.head(1)
```

```python
step1  = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop="first",handle_unknown='ignore'),[0,1,2])
],remainder='passthrough')
step3 =XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1)
pipe = Pipeline([
    ('step1',step1),
    ('step2',StandardScaler()),
    ('step3',step3)
])
pipe.fit(xtrain,ytrain)
y_pred = pipe.predict(xtest)
print('r2 score',r2_score(ytest,y_pred))
print('MAE',mean_absolute_error(ytest,y_pred))
```

```python
step1  = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop="first",handle_unknown='ignore'),[0,1,2])
],remainder='passthrough')
step3 =CatBoostRegressor()
pipe1 = Pipeline([
    ('step1',step1),
    ('step2',StandardScaler()),
    ('step3',step3)
])
pipe1.fit(xtrain,ytrain)
y_pred = pipe1.predict(xtest)
print('r2 score',r2_score(ytest,y_pred))
print('MAE',mean_absolute_error(ytest,y_pred))
```

```python
step1  = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop="first",handle_unknown='ignore'),[0,1,2])
],remainder='passthrough')
step3 =RandomForestRegressor()
pipe2 = Pipeline([
    ('step1',step1),
    ('step2',StandardScaler()),
    ('step3',step3)
])
pipe2.fit(xtrain,ytrain)
y_pred = pipe2.predict(xtest)
print('r2 score',r2_score(ytest,y_pred))
print('MAE',mean_absolute_error(ytest,y_pred))
```
